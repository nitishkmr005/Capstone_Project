```
import os
import shutil

os.mkdir('./Capstone/train')
os.mkdir('./Capstone/val')

classes = ['Atelectasis', 'Cardiomegaly', 'Effusion', 'Infiltration', 'Mass', 'Nodule', 'Pneumonia', 'Pneumothorax', 'Consolidation', 'Edema', 'Emphysema', 'Fibrosis', 'Pleural_Thickening', 'Hernia']

for subfolder in classes:
  os.mkdir('./Capstone/train/{}'.format(subfolder))
```

```
for iid,lab in (df2.values):
  File = '/content/train/{}'.format(iid)
  if '|' not in lab:
    if os.path.exists(File):
      shutil.move(File, '/content/Capstone/train/{}/'.format(lab)) 
  else:
    z=lab.split('|')
    for l in z:
      if os.path.exists(File):
        shutil.copy(File, '/content/Capstone/train/{}/'.format(l)) 
```
